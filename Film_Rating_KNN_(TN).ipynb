```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/sample_data/imdb_top_1000.csv')
df.head()
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df['Certificate'].fillna(df['Certificate'].mode()[0], inplace=True)
df['Meta_score'].fillna(df['Meta_score'].mode()[0], inplace=True)
df['Gross'].fillna(df['Gross'].mode()[0], inplace=True)
```

```python
def encode_cols(column):
  if column.dtype=='object':
    column=LabelEncoder().fit_transform(column)
  return column
```

```python
df = df.apply(encode_cols)
```

```python
x = df.iloc[:,1:-1]
y = df.iloc[:, -1]
print(x.shape)
print(y.shape)
```

```python
x_tr,x_ts,y_tr,y_ts=train_test_split(x,y,test_size=0.25)
print(x_tr.shape)
print(y_ts.shape)
```

```python
cls=KNeighborsClassifier(n_neighbors=10)
cls.fit(x_tr,y_tr)
y_prd=cls.predict(x_ts)
print("Accuracy score",accuracy_score(y_ts,y_prd)*100)
```
